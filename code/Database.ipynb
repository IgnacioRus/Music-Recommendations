```python
from config import *
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= Client_ID,
                                                           client_secret= Client_Secret_ID))
```

```python
def get_info_from_playlist_page(playlist):
    
    song_names = [track['track']['name'] for track in playlist['items'] if (track["is_local"]==False and track['track']!=None)]    
    artists_names = [', '.join([artists['name'] for artists in track['track']['artists']]) for track in playlist['items'] if (track["is_local"]==False and track['track']!=None)]
    song_uri = [track['track']['uri'] for track in playlist['items'] if (track["is_local"]==False and track['track']!=None)]
    
    playlist_dict = {'song':song_names,'artist':artists_names,'uri':song_uri}
    audio_features = sp.audio_features(song_uri)
    
    return playlist_dict, audio_features
```

```python
def get_info_from_playlist(playlist_id):
    
    playlist = sp.user_playlist_tracks("spotify", playlist_id)
    
    playlist_dict, audio_features = get_info_from_playlist_page(playlist)
    
    while playlist['next']:
        
        playlist = sp.next(playlist)
        
        new_page_play, new_page_audio = get_info_from_playlist_page(playlist)
        
        for key in playlist_dict.keys():
            playlist_dict[key].extend(new_page_play[key])
            
        audio_features.extend(new_page_audio)
    
    return playlist_dict, audio_features
        
```

```python
def many_playlists(list_of_playlists):
    
    playlist_dict, audio_features = get_info_from_playlist(list_of_playlists[0])
    i=0
    
    for playlist_id in list_of_playlists[1:]:
        
        i += 1
        print('New playlist:',i)
        
        playlist_dict_new, audio_features_new = get_info_from_playlist(playlist_id)
        
        for key in playlist_dict.keys():
            playlist_dict[key].extend(playlist_dict_new[key])
            
        audio_features.extend(audio_features_new)
    
    return playlist_dict, audio_features
```

```python
def playlists_to_df(list_of_playlists):
    
    playlist_dict, audio_features = many_playlists(list_of_playlists)
    
    table_songs = pd.DataFrame(playlist_dict)
    table_features = pd.DataFrame(audio_features)
    
    full_table = table_songs.merge(table_features,on='uri').drop_duplicates(subset=['uri'])
    
    print('Done!')
    return full_table
```
